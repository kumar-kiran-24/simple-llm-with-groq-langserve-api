```python
import os
from dotenv import load_dotenv
load_dotenv()

import openai
openai.key=os.getenv("OPENAI_API_KEY")
groq_api_key=os.getenv("GROK_API_KEY")
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.1-8b-instant",groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Transalate the following from the Engilsh to French "),
    HumanMessage(content="How are you")
]

result=model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(result)
```

using the LECL cain the components

```python
chain=model|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
genric_template="Translate the follwing into {language}"
prompt=ChatPromptTemplate([("system",genric_template),("user","{text}")])
```

```python
res=prompt.invoke({"language":"Kannada","text":"hi how are you  ?"})
```

```python
chain.invoke(res)
```

```python
chain=prompt|model|parser
chain.invoke({"language":"Kannada","text":"hello"})
```
